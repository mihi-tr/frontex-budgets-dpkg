```python
import util,re,csv
```

```python
r=util.etree_from_file("../data/raw/Budget_2011.pdf")
```

```python
data=[i.text for i in r.xpath("//text[@font=3]")]
```

```python
data=util.spliton("A-[0-9]{4}.*?",data)
```

```python
data=[reduce(util.join_text,i,[]) for i in data]
```

```python
data
```

```python
with open("../data/scraped/2011.csv","wb") as f:
    w=csv.writer(f)
    w.writerow(["item","commitment 2009","payment 2009","2010 N2","2011","2011 Draft"])
    for d in data:
        w.writerow([i.encode("utf-8") for i in d])
```
